A scatter plot is a type of plot that displays the relationship between two variables by plotting them as points on a two-dimensional plane. It is useful for identifying patterns and trends in data and for detecting outliers or unusual observations.

To generate a scatter plot using the provided code, first, you need to import the Matplotlib library and its Pyplot module. Then, you can create a scatter plot by calling the scatter function and passing the x and y data as arguments. Finally, you can add a title, xlabel, and ylabel to the plot using the title, xlabel, and ylabel functions, respectively. Here's the code to generate a scatter plot:

```python
import matplotlib.pyplot as plt
import numpy as np
np.random.seed(3)
x = 3 + np.random.normal(0, 2, 50)
y = 3 + np.random.normal(0, 2, len(x))

plt.scatter(x, y)
plt.title('Scatter Plot')
plt.xlabel('X axis')
plt.ylabel('Y axis')
plt.show()
```

This will create a scatter plot of x and y data with a title, xlabel, and ylabel. You can adjust the appearance of the plot by changing the color, size, or shape of the points, and by adding a legend or a grid.

Q3:-

The subplot() function in Matplotlib is used to create multiple plots in a single figure. It allows you to divide the figure into multiple subplots and specify the location of each subplot using the row, column, and index parameters. This function is useful for comparing multiple plots or for creating a complex figure with multiple plots.

Here's an example code to create four line plots using the subplot() function:

```python
import numpy as np
import matplotlib.pyplot as plt

# Data for line 1
x1 = np.array([0, 1, 2, 3, 4, 5])
y1 = np.array([0, 100, 200, 300, 400, 500])

# Data for line 2
x2 = np.array([0, 1, 2, 3, 4, 5])
y2 = np.array([50, 20, 40, 20, 60, 70])

# Data for line 3
x3 = np.array([0, 1, 2, 3, 4, 5])
y3 = np.array([10, 20, 30, 40, 50, 60])

# Data for line 4
x4 = np.array([0, 1, 2, 3, 4, 5])
y4 = np.array([200, 350, 250, 550, 450, 150])

# Create a figure with 4 subplots
fig, axs = plt.subplots(2, 2)

# Plot line 1 in the first subplot
axs[0, 0].plot(x1, y1)
axs[0, 0].set_title('Line 1')

# Plot line 2 in the second subplot
axs[0, 1].plot(x2, y2)
axs[0, 1].set_title('Line 2')

# Plot line 3 in the third subplot
axs[1, 0].plot(x3, y3)
axs[1, 0].set_title('Line 3')

# Plot line 4 in the fourth subplot
axs[1, 1].plot(x4, y4)
axs[1, 1].set_title('Line 4')

# Add a title to the entire figure
fig.suptitle('Four Line Plots')

# Adjust the layout of the subplots to prevent overlapping
plt.tight_layout()

# Show the figure
plt.show()
```

This code creates a figure with four line plots arranged in a 2x2 grid using the subplot() function. Each line plot is created by calling the plot() function and passing the corresponding x and y data. The set_title() function is used to add a title to each subplot. Finally, the suptitle() function is used to add a title to the entire figure, and the tight_layout() function is used to adjust the layout of the subplots to prevent overlapping.

A bar plot, also known as a bar chart, is a type of plot that displays categorical data as rectangular bars with heights or lengths proportional to the values that they represent. It is used to compare different categories or groups based on their numerical values.

Here's an example code to create a bar plot and a horizontal bar plot using the provided data:

```python
import numpy as np
import matplotlib.pyplot as plt

# Data for the bar plot
company = np.array(["Apple", "Microsoft", "Google", "AMD"])
profit = np.array([3000, 8000, 1000, 10000])

# Create a bar plot
plt.bar(company, profit)
plt.title('Company Profit')
plt.xlabel('Company')
plt.ylabel('Profit (in millions USD)')
plt.show()

# Create a horizontal bar plot
plt.barh(company, profit)
plt.title('Company Profit')
plt.xlabel('Profit (in millions USD)')
plt.ylabel('Company')
plt.show()
```

In the first part of the code, we define two arrays: company and profit, which represent the names of the companies and their corresponding profits in millions of USD. We then create a vertical bar plot using the bar() function, which takes two arrays as arguments: the x-axis values and the y-axis values. We also add a title, x-axis label, and y-axis label using the title(), xlabel(), and ylabel() functions, respectively. Finally, we show the plot using the show() function.

In the second part of the code, we create a horizontal bar plot using the barh() function instead of the bar() function. This function creates a horizontal bar plot where the y-axis and the bars are switched compared to the vertical bar plot. We also add the same title and labels as before and show the plot using the show() function.

A box plot, also known as a box-and-whisker plot, is a type of plot that displays the distribution of numerical data based on five key summary statistics: the minimum, first quartile (Q1), median, third quartile (Q3), and maximum. The box represents the interquartile range (IQR), which is the range between Q1 and Q3, while the whiskers extend to the minimum and maximum values within 1.5 times the IQR.

Box plots are used to identify the distribution, spread, and outliers of the data, and to compare the distributions of different groups or datasets.

Here's an example code to create a box plot using the provided data:

```python
import numpy as np
import matplotlib.pyplot as plt

# Data for the box plot
box1 = np.random.normal(100, 10, 200)
box2 = np.random.normal(90, 20, 200)

# Combine the data into a list
data = [box1, box2]

# Create a box plot
plt.boxplot(data)
plt.title('Box Plot')
plt.xticks([1, 2], ['Box 1', 'Box 2'])
plt.ylabel('Value')
plt.show()
```

In the code, we first generate two sets of random data box1 and box2 using the np.random.normal() function. We then combine the data into a list data, which is required for creating a box plot with multiple boxes.

We create the box plot using the boxplot() function, which takes the data as input. We also add a title, x-axis labels, and a y-axis label using the title(), xticks(), and ylabel() functions, respectively. Finally, we show the plot using the show() function.

The resulting plot will display a box for each dataset, with whiskers extending to the minimum and maximum values within 1.5 times the IQR. The median of each dataset will be represented by a line inside the box. Any outliers beyond the whiskers will be plotted as individual points.
