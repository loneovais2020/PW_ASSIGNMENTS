Q2:-

 Here's an example Python function that estimates the population mean using a sample mean and standard deviation:

```python
def estimate_population_mean(sample_mean, sample_std, sample_size):
    standard_error = sample_std / (sample_size ** 0.5)
    z_score = 1.96 # for 95% confidence interval
    lower_bound = sample_mean - z_score * standard_error
    upper_bound = sample_mean + z_score * standard_error
    return (lower_bound, upper_bound)

estimate_population_mean(102,4.5,38)

```

This function takes in the sample mean, sample standard deviation, and sample size as input, and returns a tuple containing the lower and upper bounds of the 95% confidence interval for the population mean.

Q5:-

 Here's an example Python script that conducts a hypothesis test on the difference between two population means:

```python
import numpy as np
from scipy.stats import ttest_ind

# Generate two samples from two populations
sample1 = np.random.normal(10, 2, size=50)
sample2 = np.random.normal(12, 2, size=50)

# Calculate the sample means and standard deviations
mean1, mean2 = np.mean(sample1), np.mean(sample2)
std1, std2 = np.std(sample1), np.std(sample2)

# Calculate the t-statistic and p-value
t_stat, p_value = ttest_ind(sample1, sample2)

# Print the results
print("Sample 1 mean:", mean1)
print("Sample 2 mean:", mean2)
print("Sample 1 std:", std1)
print("Sample 2 std:", std2)
print("t-statistic:", t_stat)
print("p-value:", p_value)
```

This script generates two samples from two populations, calculates the sample means and standard deviations, and then uses the t-test to test whether the difference between the means is statistically significant.

 Here's an example Python code to generate a Student's t-distribution plot using matplotlib with 10 degrees of freedom:-

```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import t

df = 10
x = np.linspace(t.ppf(0.01, df), t.ppf(0.99, df), 100)
plt.plot(x, t.pdf(x, df))
plt.title("Student's t-Distribution with df=10")
plt.xlabel('t-value')
plt.ylabel('Probability Density')
plt.show()
```

Q14:-

 A researcher hypothesizes that a new drug will decrease blood pressure by 10 mmHg. They conduct a clinical trial with 100 patients and find that the sample mean decrease in blood pressure is 8 mmHg with a standard deviation of 3 mmHg. Test the hypothesis with a significance level of 0.05.

To test the hypothesis, we can use a one-sample t-test. The null hypothesis is that the mean decrease in blood pressure is equal to or greater than 10 mmHg, while the alternative hypothesis is that the mean decrease is less than 10 mmHg.

Let's set the significance level at 0.05.

Null hypothesis: 𝜇 >= 10

Alternative hypothesis: 𝜇 < 10


The test statistic can be calculated as follows:

t = (x̄ - 𝜇) / (s / sqrt(n))

t = (8 - 10) / (3 / sqrt(100))

t = -6.67


The degrees of freedom for this test is 99 (n-1).

Using a t-distribution table or calculator, the p-value for a one-tailed test with 99 degrees of freedom and t = -6.67 is less than 0.0001.

Since the p-value is less than the significance level of 0.05, we reject the null hypothesis. There is evidence to support the claim that the new drug decreases blood pressure by less than 10 mmHg.

Therefore, we can conclude that the new drug is not effective in reducing blood pressure by 10 mmHg.
