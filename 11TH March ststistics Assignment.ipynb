Q1:-

 The main difference between a t-test and a z-test is the type of data they can be used for. A z-test is used when the population standard deviation is known, while a t-test is used when the population standard deviation is unknown and needs to be estimated from the sample.

For example, if a researcher wants to test if the mean height of a population is significantly different from a given value and the population standard deviation is known, a z-test would be appropriate. On the other hand, if the population standard deviation is unknown, the researcher could use a t-test to estimate the standard deviation from the sample and test the hypothesis.

 A one-tailed test is a hypothesis test where the alternative hypothesis specifies the direction of the effect, either positive or negative. A two-tailed test is a hypothesis test where the alternative hypothesis does not specify the direction of the effect, but only that there is a difference or relationship between the variables being tested.

For example, in a one-tailed test, a researcher might want to test whether a new drug treatment increases patients' survival rate compared to the control group. The null hypothesis would be that the survival rate is the same in both groups, and the alternative hypothesis would be that the survival rate in the treatment group is higher. In a two-tailed test, the researcher might want to test whether there is a difference in IQ scores between two groups, but without specifying whether one group has a higher or lower score than the other.



Q3:-

 Type 1 and Type 2 errors are two possible errors that can occur in hypothesis testing. A Type 1 error occurs when the null hypothesis is rejected when it is actually true. This is also known as a false positive. A Type 2 error occurs when the null hypothesis is accepted when it is actually false. This is also known as a false negative.

For example, suppose a researcher wants to test whether a new drug is effective in treating a disease. A Type 1 error would occur if the researcher concludes that the drug is effective when it is actually not. This could lead to patients being given a drug that is not effective and potentially harmful. A Type 2 error would occur if the researcher concludes that the drug is not effective when it is actually effective. This could lead to patients not receiving a treatment that could be helpful.

 Bayes's theorem is a mathematical formula that describes the probability of an event based on prior knowledge of conditions that might be related to the event. The formula is:

P(A|B) = P(B|A) * P(A) / P(B)

Where P(A|B) is the probability of A given B, P(B|A) is the probability of B given A, P(A) is the prior probability of A, and P(B) is the prior probability of B.

For example, suppose a doctor wants to know the probability that a patient has a particular disease given that the patient has a positive test result. The doctor knows that the disease is rare, with a prevalence of 1%, and that the test has a false positive rate of 5% and a false negative rate of 2%. Using Bayes's theorem, the doctor can calculate the probability of the patient having the disease given a positive test result.

P(Disease|Positive Test) = P(Positive Test|Disease) * P(Disease) / P(Positive Test)

= 0.98 * 0.01 / (0.98 * 0.01 + 0.05 * 0.99)

= 0.163

This means that the probability of the patient having the disease given a positive test result is 16.3%.

 A confidence interval is a range of values that is likely to contain the true value of a population parameter with a certain level of confidence. It is often used in statistical inference to estimate population parameters based on a sample of data.

To calculate a confidence interval, one needs to specify the level of confidence, the sample mean, and the sample standard deviation (or the sample size and the population standard deviation). The formula for a confidence interval for the population mean is:

Confidence interval = sample mean ± (z-score or t-score * standard error)

where the standard error is the standard deviation of the sampling distribution of the mean, which is equal to the population standard deviation divided by the square root of the sample size.

For example, suppose a sample of 100 students is taken from a university, and the sample mean GPA is 3.5 with a sample standard deviation of 0.5. We want to calculate a 95% confidence interval for the population mean GPA. The critical value for a 95% confidence interval with 99 degrees of freedom (n-1) is 1.984 (from a t-distribution table). The standard error is calculated as 0.05 (sample standard deviation) / sqrt(100) = 0.005.

The confidence interval is then:

3.5 ± (1.984 * 0.005) = (3.49, 3.51)

This means that we can be 95% confident that the true population mean GPA is between 3.49 and 3.51.

Q10:-

 In a study of the effectiveness of a new weight loss drug, a sample of 50 participants lost an average
of 6 pounds with a standard deviation of 2.5 pounds. Conduct a hypothesis test to determine if the drug is
significantly effective at a 95% confidence level using a t-test.

To test the hypothesis, we can set up the following:

Null hypothesis (H0): The mean weight loss is not significantly different from 0.
Alternative hypothesis (HA): The mean weight loss is significantly different from 0.

We will use a two-tailed t-test since we are interested in whether the weight loss is significantly different from both positive and negative values.

The t-statistic can be calculated as:

t = (x̄ - μ) / (s / √n)
where x̄ = 6, μ = 0, s = 2.5, and n = 50

Plugging in the values, we get:

t = (6 - 0) / (2.5 / √50) = 12

Using a t-table or calculator with 49 degrees of freedom (n-1), we find that the p-value is less than 0.001.

Since the p-value is less than our significance level of 0.05, we reject the null hypothesis and conclude that the mean weight loss is significantly different from 0.

Therefore, we can conclude that the weight loss drug is effective at a 95% confidence level.
